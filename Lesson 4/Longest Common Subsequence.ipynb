### 3. Come up with a correct solution for the problem. State it in plain English.

#### Recursive Solution


1. Create two counters `idx1` and `idx2` starting at 0. Our recursive function will compute the LCS of `seq1[idx1:]` and `seq2[idx2:]`


2. If `seq1[idx1]` and `seq2[idx2]` are equal, then this character belongs to the LCS of `seq1[idx1:]` and `seq2[idx2:]` (why?). Further the length this is LCS is one more than LCS of `seq1[idx1+1:]` and  `seq2[idx2+1:]`

<img src="https://i.imgur.com/um7LDiX.png" width="400">

3. If not, then the LCS of `seq1[idx1:]` and `seq2[idx2:]` is the longer one among the LCS of `seq1[idx1+1:], seq2[idx2:]` and the LCS of `seq1[idx1:]`, `seq2[idx2+1:]`

<img src="https://i.imgur.com/DRanmOy.png" width="360">

5. If either `seq1[idx1:]` or `seq2[idx2:]` is empty, then their LCS is empty.

###  4. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
def lcs_recursive(seq1, seq2, idx1=0, idx2= 0):
    # base case:
    if idx1 == len(seq1) or idx2 == len(seq2):
        return 0
    
    # recursive call:
    elif seq1[idx1] == seq2[idx2]:
        # do something before the recursive call (add 1)
        return 1 + lcs_recursive(seq1, seq2, idx1+1, idx2+1)
    else:
        return max(lcs_recursive(seq1, seq2, idx1+1, idx2), lcs_recursive(seq1, seq2, idx1, idx2+1))
```

### Memoization (Top-Down):

Same recursive function is used, but results of each subtree is stored inside a dict.

if a subtree that has already been solved occurs again, the dict can call it in constant time O(1), instead of solving that subtree again.

If looked into a matrix, the table is filled from bottom-up; the memoized recursive function is top-down.

<img src="https://i.imgur.com/z5m36m8.png" width="360">


<img src="https://i.imgur.com/n8ZgBYj.png" width="500">

#### https://youtu.be/sSno9rV8Rhg

```python
def lcs_recursive(seq1, seq2, idx1=0, idx2= 0):
    # base case:
    if idx1 == len(seq1) or idx2 == len(seq2):
        return 0
    
    elif seq1[idx1] == seq2[idx2]:
        return 1 + lcs_recursive(seq1, seq2, idx1+1, idx2+1)
    
    else:
        return max(lcs_recursive(seq1, seq2, idx1+1, idx2), lcs_recursive(seq1, seq2, idx1, idx2+1))
```

```python
def lcs_memo(seq1, seq2):
    memo = {}
    def recur(idx1=0, idx2=0):

        # calling a solution of a subtree in O(1) time if that subtree has already been called
        key = (idx1, idx2) # an individual key made up of (idx1, idx2) for each subtree
        if key in memo:
            return memo[key]
        
        # the lcs_recur func is just repeated here
        # a KEY DIFFERENCE is that each ans of a subtree is stored into the dict with its allocated key
        elif idx1 == len(seq1) or idx2 == len(seq2):
            memo[key] = 0
        elif seq1[idx1] == seq2[idx2]:
            memo[key] = 1 + recur(idx1+1, idx2+1)
        else:
            memo[key] = max(recur(idx1+1, idx2), recur(idx1, idx2+1))

        # the main tree's/head's value is returned -> memo[ (0,0) ] = lcs
        return memo[key]
    
    return recur()
```

### Dynamic Programming (Bottom-Up):
DP uses a bottom-up approch, but the matrix is filled from top-down.

DP uses iteration instead of recursion which makes it a bit faster than Memoized Recursion

<img src="https://i.imgur.com/SAsEol6.png">


#### https://youtu.be/sSno9rV8Rhg

```python
# Creating a matrix

s1 = 'bd'
s2 = 'abcd'
tab = [ ([None]*(len(s2)+1))  for i in range(len(s1)+1)]
print(tab,'\n')

print(s1+' x '+s2)
s=''
for i in range(len(tab)):
    for j in range(len(tab[0])):
        s += str(tab[i][j]) + '\t|\t'
    print(s, "\n")
    s=''
```

```python
def lcs_dynamic(seq1, seq2):

    # Creating the matrix
    table = [([0]*(len(seq2)+1)) for i in range(len(seq1)+1)]
    
    # A helper function that fills up a spot in the matrix
    def fill_table(i, j):
        if seq1[i-1] == seq2[j-1]:
            table[i][j] = 1 + table[i-1][j-1]
        else:
            table[i][j] = max(table[i-1][j], table[i][j-1])


    # The 2 loops fills up all the spots in the matrix
    for i in range(1, len(table)):
        for j in range(1, len(table[0])):
            fill_table(i, j)

    # The last element in the matrix holds value of LCS
    return table[-1][-1]
```
