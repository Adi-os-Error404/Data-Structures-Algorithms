# I : The Greedy Algorithm

https://youtu.be/oTTzNMHM05I

```python
def greedy_max_profit(profits, weights, capacity):

    # sorted() returns a list, where as .sort() returns nothing but changed the orignal list
    # zip is used to get elements from 2 lists at the same time
    # enumerate an be used with zip to get the index of those elements
    ratio_list = [(i, p/f) for i, (p,f) in enumerate( zip(profits, weights) )]

    # key take a function (ex: key=len, key=helper_func); 
    # the lambda func is being used to use the second el of the tuple in ratio_list to sort, and reverse=True is sorting in descending order
    ratio_list.sort(key= lambda tup: tup[1], reverse=True)


    max_profit = 0
    cur_cap = 0
    i = 0
    while cur_cap <= capacity:
        if (i == (len(ratio_list)-1)) or ((cur_cap+weights[ratio_list[i][0]]) > capacity):
            break

        max_profit += profits[ratio_list[i][0]]
        cur_cap += weights[ratio_list[i][0]]
        i+=1
        
    return max_profit
```
